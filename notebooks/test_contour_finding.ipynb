```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from src.data.datacollection import get_box_positions, extract_moves_from_image 
 
```

## Testing that we can find the contours of the image

```python

# fix seed of cv2 
np.random.seed(42)
cv2.setRNGSeed(42)

# Load the image
image_path = "notebooks/table2.png"
image = cv2.imread(image_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

rects = get_box_positions(image) 

# draw the rectangles on the image 
output_image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
for i, rect in enumerate(rects): 
    x, y, w, h = rect 
    cv2.rectangle(output_image_rgb, (x, y), (x+w, y+h), (0, 255, 0), 2)
    # draw the index of the box 
    cv2.putText(output_image_rgb, str(i), (rect[0], rect[1]), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

plt.figure(figsize=(10, 10))
plt.imshow(output_image_rgb)
plt.title("Detected Boxes")
plt.axis("off")
plt.show()
```
